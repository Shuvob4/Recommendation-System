## Content-Based Filtering 
Content-based filtering is a technique used in recommendation systems that analyzes the characteristics or properties of an item to recommend similar items to users. In content-based filtering, the system learns from the attributes of the items that the user has previously interacted with, and then recommends items with similar attributes. Content based filtering can be classified as Classifier-based and Neighbourhood-based. Both method used user interaction data and difference is how they use the data in each method. For Classifier-based method, the system learns a classifier model from the features of items that users have interacted with in the past. The model is then used to predict the likelihood that a user will interact with a new item, based on its features. While, In neighborhood-based approach, the system identifies a set of items that are similar to the items that the user has interacted with in the past. The system then recommends items from this set to the user. advantage of classifier-based content-based filtering is that it can handle more complex item features than simple similarity metrics. It also has the potential to handle new items with no prior user interactions, as long as the item features are well defined.

**Types of Content-Based Filtering :**
1. Classifier-Based
2. Neighbourhood-Based

**Steps in Content Based Filtering :**
1. Item Representation
2. User Profile Creation
3. Similarity Computation
4. Recommendation Generation

**Advantage :**
* It does not require explicit feedback from the user, such as ratings or reviews. Instead, it relies on the inherent characteristics of the items to make recommendations. 

**Drawback :**
* It may not be able to recommend items that are outside the user's existing interests, as it relies on the user's past interactions to create their profile.

## Collaborative Filtering 

Collaborative filtering is a technique used in recommendation systems that analyzes the past behavior or preferences of a group of users to recommend items to individual users. In collaborative filtering, the system learns from the behavior of other users with similar preferences to generate recommendations for a target user.  It does not need to analyze the content of the items. Instead, it relies on the relationship between users and items, which are typically encoded in a rating feedback matrix with each element representing a specific user rating on a specific item. Collaborative filtering is based on the assumption that users with similar tastes for previous items will have similar preferences for new items, so the algorithm recommends items ranked highly by users deemed similar to the current user. Such algorithms fall into two main categories, Neighbourhood-based and Model-based. 

Neighborhood-based collaborative filtering is a popular approach to collaborative filtering that identifies a set of similar users to a target user, and then generates recommendations based on the behavior of those similar users.

Model-based collaborative filtering is an approach to collaborative filtering that uses machine learning algorithms to learn the underlying patterns and relationships in the user-item matrix, and then generates recommendations based on those patterns and relationships.

**Types of Collaborative Filtering :**
1. Neighborhood-Based Collaborative Filtering
2. Model-Based Collaborative Filtering

**Steps in Collaborative Filtering :**
1. User-item matrix
2. Similarity computation
3. Neighborhood identification
4. Rating prediction
5. Recommendation Generation

**Advantage :**
* Can handle the **'cold start'** problem, where there is not enough data on the user's past interactions, as it relies on the behavior of similar users.
* Can recommend items that are outside the user's existing interests, as it learns from the behavior of other users with similar preferences.
*  Relatively simple to implement and can generate accurate recommendations with smaller data sets.

**Drawback :**
* Can be computationally expensive as it requires training and optimizing machine learning algorithms on a large user-item matrix. 
*  Suffer from the **'overfitting'** problem, where the model becomes too complex and fits too closely to the training data, leading to poor generalization to new users and items.
* Suffer from the **'sparsity'** problem, where the user-item matrix has a large number of empty entries, making it difficult to identify similar users.
* Suffer from the **'gray sheep'** problem, where some users have unique preferences that do not match well with any other users, making it difficult to find similar users. 
* Requires a significant amount of data to train the machine learning algorithms effectively.

## Hybrid Filtering

Hybrid filtering is an approach to recommendation systems that combines two or more different types of recommendation algorithms to provide more accurate and personalized recommendations to users.

Hybrid Filtering can be implemented in several ways, such as:

* By making content-based and collaborative-based predictions separately and then combining them.
* By adding content-based capabilities to a collaborative-based approach (and vice versa) or
* By combining the approaches into one model.

Some hybridization techniques include:
* **Weighted**
* **Switching**
* **Mixed**
* **Feature Combination**
* **Feature Augmentation**

## Generic Problem Faced in Recommendation Systems
### Cold Start Problem
The cold start problem is a challenge that arises in recommendation systems when there is not enough data available to make accurate recommendations for new users or new items that have just been added to the system.

For new users, the system does not have enough historical data on their interactions with the system, such as their ratings, purchases, or clicks, to understand their preferences and make relevant recommendations. For new items, the system does not have enough information about the item's characteristics, such as its genre, director, or actors, to recommend it to users who might be interested in it.

There are several ways to address this issue in RS, such as
* **Content-based filtering**
* **Knowledge-based recommendations**
* **Hybrid filtering**
* **Feature Combination**
* **Feature Augmentation**
* **Active learning**


### Sparsity Problem
The sparsity problem is a common challenge in recommendation systems that arises when the available data on user-item interactions is very sparse, i.e., there are very few ratings or interactions between users and items. Sparsity can make it difficult for recommendation algorithms to accurately capture the preferences of users and make relevant recommendations.

Sparsity is a problem for many recommendation algorithms, but it is particularly challenging for collaborative filtering, which relies on the similarity between users or items to make recommendations. When the data is very sparse, it becomes difficult to compute accurate similarity measures, and the recommendations may be biased or unreliable.

There are several ways to address the sparsity problem in recommendation systems, including:
* **Data augmentation**
* **Regularization**
* **Matrix factorization**
* **Neighborhood-based methods**


### Data quality and quantity problem
Data quality and quantity are critical factors in building effective recommendation systems. The quality of the data used to train the recommendation system directly impacts the accuracy and relevance of the recommendations provided. The following are some of the problems related to data quality and quantity in recommendation systems:

* **Data Bias**
* **Incomplete Data**
* **Noise**

To overcome these challenges, data pre-processing techniques such as data cleaning, feature engineering, and normalization can be employed to improve data quality. Also, data augmentation techniques such as incorporating social network data, item descriptions, or user demographics can be employed to address data sparsity and improve the quantity of data available.


### Scalability Problem

Scalability is a critical challenge in building recommendation systems, particularly as the amount of data and the number of users and items increase. A recommendation system that works well with a small dataset may not be scalable to a larger dataset, which can make it challenging to provide real-time or near-real-time recommendations. The following are some of the scalability problems in recommendation systems:

* **Computation time**
* **Memory Limitations**
* **Distributed data**

To address scalability challenges, recommendation systems can employ various techniques such as:
* **Parallel computing**
* **Sampling techniques**
* **Distributed computing**

## Applying weighted average Function to predict rating based recommendation system (Simple Recommendation Sytem based on statistical Formula of weighted average)

They offer generalized recommendations to every user, based on products popularity. The System recommends the same products to users with similar demographic features. Since each user is different , this approach is considered to be too simple. The basic idea behind this system is that products that are more popular and critically acclaimed will have a higher probability of being liked by the average consumers.

Before getting started with this -
* we need a metric to score or rate product
* Calculate the score for every product
* Sort the scores and recommend the best rated product to the users.

We can use the average ratings of the product as the score but using this won't be fair enough since a product with 4.75 average rating and only 2 votes cannot be considered better than the product with 4.6 as as average rating but 40 votes. So, We'll be using weighted rating (wr) which is given as :-

<img src="wr.png">

where,

* v is the number of rating received for the product;
* m is the minimum ratings required to be listed in the chart;
* R is the average rating of the product; And
* C is the mean vote across the whole report

We already have v(no_of_rating_received) and R (rating) and C can be calculated as

```python
# creating a new dataframe with some features of metadata
product_ = meta_data[['productid','title', 'brand', 'rank']]

#merging product related dataframe together
product_details = pd.merge(products, product_, on='productid')
product_details.head()
```

```python
#checking for duplicate products in the dataset 
print(sum(product_details.duplicated('productid')))
```

We will remove the duplicates from the dataset first.

```python
#deleting duplicates value from dataset while keeping the first entry
product_details.drop_duplicates(subset=['productid'], keep='first', inplace=True)

#checing if deletion works or not
print(sum(product_details.duplicated('productid')))
```

As we are able to remove duplicacy in our data, We will apply weighted function on the above dataset. To do so, we need to calculate the unknowns of the above equation

```python
# calculating the mean vote accross the whole dataset
C = product_details['rating'].mean()
C
```

```python
product_details.describe()
```

From the above calculation, we can see that value of C is 4.04 and 50% value falls between the range of the C. Next we will determine an appropriate value for m, the minimum ratings required to be listed in the chart. We will use 90th percentile as our cutoff. In other words, for a product to feature in the charts.

```python
#calculating the minimum no of ratings requires to be featured in the final dataset
m = product_details['no_of_rating_received'].quantile(0.9)
m
```

```python
# making the final dataset based on the above value of m
q_products = product_details.copy().loc[product_details['no_of_rating_received'] >= m]
q_products.shape
```

We can see that, from a dataset of 32486 items, it has come to 3296 items, In other words, among all the products these are the products that sells most.

```python
#defining the weight function
def weight_rating_calculation(df, m=m, C=C):
    v = df['no_of_rating_received']
    R = df['rating']
    return (v/(v+m) * R) + (m/(m+v) * C)
```

```python
#Applying the weighted average function to determine the weighted rating of each products in final dataset
q_products['weighted_rating'] = q_products.apply(weight_rating_calculation, axis=1)

#sort products based on calculated score above
q_products = q_products.sort_values('weighted_rating', ascending=False)

#displaying the top 30 items based on weighted rating
q_products.head(30)
```

```python
#calculating mean & mode for both columns
rating_mean = q_products['rating'].mean()
rating_mode = q_products.mode()['rating'][0]
rating_std = 0.597669 #from the table above

weighted_rating_mean = q_products['weighted_rating'].mean()
weighted_rating_mode = q_products.mode()['weighted_rating'][0]
weighted_rating_std = 0.412228 #from the table above


#calculating pearson  first coefficient of skewness
rating_pfc = (rating_mean - rating_mode) / rating_std
weighted_rating_pfc = (weighted_rating_mean - weighted_rating_mode) /weighted_rating_std

#calculating pearson second coefficient of skewness
rating_psc = (3 * (rating_mean - rating_mode)) / rating_std
weighted_rating_psc = (3 * (weighted_rating_mean - weighted_rating_mode)) /weighted_rating_std

print('Pearsons first coefficient of skewness : ')
print('For Rating Column : ', rating_pfc)
print('For Weighted Rating Column : ', weighted_rating_pfc)

print('-' * 50)
print('Pearsons Second coefficient of skewness : ')
print('For Rating Column : ', rating_psc)
print('For Weighted Rating Column : ', weighted_rating_psc)
```

From the above calculation, we can say that, rating column is negatively skewed while weighted rating column is positively skewed. 

```python
#calculating threshold value for both column
threshold_value_for_rating = np.percentile(q_products['rating'], 25)
threshold_value_for_weighted_rating = np.percentile(q_products['weighted_rating'], 75)
```

```python
# Convert ranking values to binary values based on the threshold
q_products['rating_binary'] = np.where(q_products['rating'] >= threshold_value_for_rating, 1, 0)
q_products['weighted_rating_binary'] = np.where(q_products['weighted_rating'] >= threshold_value_for_weighted_rating, 1, 0)

# Calculate precision using precision_score from scikit-learn
precision_DMWA = precision_score(q_products['rating_binary'], q_products['weighted_rating_binary'])
print('\nPrecision Score of Demograpic Model Using Weighted Average = {} \n'.format(precision_DMWA))
```

In the above dataframe, we can see that userid(index) is the object datatype, We will replace the userid by numbers starting from 0 to 1398 for all userid so that, the index is of integer type and represents a userid in the same format.

```python
#changing userid from object format to int format 
rating_matrix_final['user_index'] = np.arange(0, rating_matrix_final.shape[0])
rating_matrix_final.set_index(['user_index'], inplace=True)

#checking the dataframe
rating_matrix_final.head()
```

```python
#defining a function to get similar type of users
def finding_similar_users(user_index, matrix):
    #1st param: it's the user index based on which we will try to find similar user
    #2nd param: it's a matrix which is user interations with product
    
    #defining empty list for similarity
    similarity = []
    
    #checking every user in the matrix
    #.shape[0] gives the number of rows, in our case: total number of users 
    for user in range(0, matrix.shape[0]):
        
        #using cosine similarity to find similarity between given userid and every user in df
        cos_sim = cosine_similarity([matrix.loc[user_index]], [matrix.loc[user]])

        #Appending the user and the corresponding similarity score with userid as a tuple
        similarity.append((user,cos_sim))
        
    similarity.sort(key=lambda x: x[1], reverse=True)
    #Extracting the user from each tuple in the sorted list
    most_similar_users = [tup[0] for tup in similarity]
    #Extracting the similarity score from each tuple in the sorted list
    similarity_score = [tup[1] for tup in similarity] 
   
    #Remove the original user and its similarity score and keep only other similar users 
    most_similar_users.remove(user_index)
    similarity_score.remove(similarity_score[0])
       
    return most_similar_users, similarity_score
```

As we have get users with similar kind of interest of products, now we will write a function for getting product recommendation

```python
#defining the recommendation finction for getting product recommendation by using similar users preference
def get_product_recommendations(user_index, no_of_products, matrix):
    #1st param: it's the user index based on which we will get similar users
    #2nd Param: Number of products that will be recommended
    #3rd Param: it's a matrix which is user interations with product

    #getting most similar users
    most_sim_users = finding_similar_users(user_index, matrix)[0]
    
    #getting productid with the given userid has interacted
    product_ids = set(list(matrix.columns[np.where(matrix.loc[user_index] > 0)]))
    
    #declaring an emplty list for storing recommended products
    recommended_products = []
    
    #creating a copy of product_ids
    interaction_observed_matrix = product_ids.copy()
    
    #looping through most_similar_users to get recommended products
    for sim_user in most_sim_users:
        if len(recommended_products) < no_of_products:
            #finding top n products which have been rated by similar users
            sim_user_product_ids = set(list(matrix.columns[np.where(matrix.loc[sim_user] > 0 )]))
            recommended_products.extend(list(sim_user_product_ids.difference(interaction_observed_matrix)))
            interaction_observed_matrix = interaction_observed_matrix.union(sim_user_product_ids)
        else:
            break
    return recommended_products[: no_of_products]
```

```python
#recommending products based on similarity user prefrence for userid 210
get_product_recommendations(210, 5, rating_matrix_final)
```

```python
#recommending products based on similarity user prefrence for userid 727
get_product_recommendations(727, 5, rating_matrix_final)
```

From the above output we can see that, just like before our model is predicting products as per similar user prefences as collaborative filtering is based on the assumption of people who share similarity in previously browsed products are most likely to buy items from that similar user bought product which he hasn't bought yet.

# Model Based Collaborative Filtering

For our model based collaborative filtering we will use single value decomposition as our matrix is large and highly sparse.

```python
#Singular Value Decomposition for sparse matrix
#k is the number of latent features
u, s, vt = svds(rating_matrix_final, k = 50) 

#Constructing diagonal array in SVD
sigma = np.diag(s)
```

```python
#checking the shape of our matrix
print('U Matix Shape : ', u.shape)
print('Sigma Matix Shape : ', sigma.shape)
print('VT Matix Shape : ', vt.shape)
```

We will generate original matrix using u, sigma and vt matrices. The resulting matrix would be the predicted ratings for all users and products.

```python
#creating users predicted rating matrix
users_predicted_ratings_matrix = np.dot(np.dot(u, sigma), vt)

#predicted rating dataframe
predicted_rating_df = pd.DataFrame(abs(users_predicted_ratings_matrix), columns=rating_matrix_final.columns)

#checking newly created dataframe
predicted_rating_df.head()
```

```python
#defining a method for recommending items with the highest predicted ratings
def item_recommendation(user_index, matrix, dataframe, no_of_recommended_items):
    #1st param: user index based on which we will recommended product
    #2nd param: user predicted rating matrix
    #3rd param: dataframe with predicted ratings
    #4th param: how many products will be recommended 
    
    #getting and sorting users rating from actual and predicted matrix
    user_actual_rating_sorted = matrix.loc[user_index].sort_values(ascending=False)
    user_predicted_rating_sorted = dataframe.loc[user_index].sort_values(ascending=False)

    #creating new dataframe with actual and predicted ratings
    temp_df = pd.concat([user_actual_rating_sorted, user_predicted_rating_sorted], axis=1)
    temp_df.index.name = 'recommended_products'
    temp_df.columns = ['actual_ratings', 'predicted_ratings']
    
    #filtering df where actual rating = 0 means that user has not interacted 
    temp_df = temp_df.loc[temp_df.actual_ratings == 0]
    
    #recommending products with top predicted reatings
    temp_df = temp_df.sort_values(by = 'predicted_ratings', ascending=False)
    print(f'Recommended Products for userid {user_index} : \n')
    print(temp_df['predicted_ratings'].head(no_of_recommended_items))
```

From the above figures we are not quiet sure how to determine the threshold value for cutoff point, so we will apply the pearson coefficient score for better understanding of the skewness of our dataframe.

```python
#calculating mean & mode for both columns
actual_rating_mean_cfsvd = rmse_CFSVD['actual_ratings_average'].mean()
actual_rating_mode_cfsvd = rmse_CFSVD.mode()['actual_ratings_average'][0]
actual_rating_std_cfsvd = 0.127618 #from the table above

predicted_rating_mean_cfsvd = rmse_CFSVD['predicted_rating_average'].mean()
predicted_rating_mode_cfsvd = rmse_CFSVD.mode()['predicted_rating_average'][0]
predicted_rating_std_cfsvd = 0.016300 #from the table above


#calculating pearson  first coefficient of skewness
actual_rating_pfc = (actual_rating_mean_cfsvd - actual_rating_mode_cfsvd) / actual_rating_std_cfsvd
predicted_rating_pfc = (predicted_rating_mean_cfsvd - predicted_rating_mode_cfsvd) /predicted_rating_std_cfsvd

#calculating pearson second coefficient of skewness
actual_rating_psc = (3 * (actual_rating_mean_cfsvd - actual_rating_mode_cfsvd)) / actual_rating_std_cfsvd
predicted_rating_psc = (3 * (predicted_rating_mean_cfsvd - predicted_rating_mode_cfsvd)) /predicted_rating_std_cfsvd

print('Pearsons first coefficient of skewness : ')
print('For Actual Rating Column : ', actual_rating_pfc)
print('For Predicted Rating Column : ', predicted_rating_pfc)

print('-' * 50)
print('Pearsons Second coefficient of skewness : ')
print('For Actual Rating Column : ', actual_rating_psc)
print('For Predicted Rating Column : ', predicted_rating_psc)
```

From the above calculation we can see how our data is skewed.

**Actual Rating Column**
As both coefficient value is between 0.0 and 0.5, thus it righly skewed

**Predicted Rating Column**
Here, first coefficient value suggest that data is positively skewed 
however, second coefficient value suggest that data is highly skewed.
For our purpose, we will use the first coefficient value.

```python
#calculating threshold value for both column
threshold_value_for_actual_rating = np.percentile(rmse_CFSVD['actual_ratings_average'], 75)
threshold_value_for_predicted_rating = np.percentile(rmse_CFSVD['predicted_rating_average'], 75)
```

```python
# Convert ranking values to binary values based on the threshold
rmse_CFSVD['actual_rating_binary'] = np.where(rmse_CFSVD['actual_ratings_average'] >= threshold_value_for_actual_rating, 1, 0)
rmse_CFSVD['predicted_rating_binary'] = np.where(rmse_CFSVD['predicted_rating_average'] >= threshold_value_for_predicted_rating, 1, 0)

# Calculate precision using precision_score from scikit-learn
precision_CFSVD = precision_score(rmse_CFSVD['actual_rating_binary'], rmse_CFSVD['predicted_rating_binary'])
print('\nPrecision Score of Model Based Collaborative Filtering using Single Value Decomposition = {} \n'.format(precision_CFSVD))
```
